Dikarenakan kedua nilai mempunyai nilai perbandingan yang hampir sama, maka untuk menentukan Item yang mempunyai nilai yang sama, akan digunakan sistem random selection dengan peluang yang berdasarkan dari total jumlah mereka.

```
np.random.seed(42)

# =========================
# Cake dan Juice
# =========================

total_cake = df[df['Item'] == 'Cake']['Quantity'].sum()
total_juice = df[df['Item'] == 'Juice']['Quantity'].sum()

total_cake_juice = total_cake + total_juice

prob_cake = total_cake / total_cake_juice
prob_juice = total_juice / total_cake_juice

mask_3 = (df['Item'].isnull() & (df['Price Per Unit'] == 3.0))

df.loc[mask_3, 'Item'] = np.random.choice(
    ['Cake', 'Juice'],
    size=mask_3.sum(),
    p=[prob_cake, prob_juice]
)


# =========================
# Smoothie dan Sandwich
# =========================

total_smoothie = df[df['Item'] == 'Smoothie']['Quantity'].sum()
total_sandwich = df[df['Item'] == 'Sandwich']['Quantity'].sum()

total_smoothie_sandwich = total_smoothie + total_sandwich

prob_smoothie = total_smoothie / total_smoothie_sandwich
prob_sandwich = total_sandwich / total_smoothie_sandwich

mask_4 = (df['Item'].isnull() & (df['Price Per Unit'] == 4.0))

df.loc[mask_4, 'Item'] = np.random.choice(
    ['Smoothie', 'Sandwich'],
    size=mask_4.sum(),
    p=[prob_smoothie, prob_sandwich]
)
```

Begitu juga dengan total spent 9.0 dan 16.0 dikarenakan tidak ada perkalian yang bisa mencapai angka total spent tersebut.

```
# ======================================================
# Mengisi Item, Quantity, dan Price Per Unit yang kosong
# berdasarkan Total Spent tertentu: 25.0, 9.0, dan 16.0
# ======================================================

np.random.seed(42)

base_mask = (
    df['Item'].isna() &
    df['Quantity'].isna() &
    df['Price Per Unit'].isna() &
    df['Total Spent'].notna()
)

# ======================================================
# Total Spent = 25.0
# Quantity = 5, Price Per Unit = 5.0, Item = Salad
# ======================================================

mask_25 = base_mask & (df['Total Spent'] == 25.0)

df.loc[mask_25, 'Quantity'] = 5
df.loc[mask_25, 'Price Per Unit'] = 5.0
df.loc[mask_25, 'Item'] = 'Salad'


# ======================================================
# Total Spent = 9.0
# Quantity = 3, Price Per Unit = 3.0
# Item = Cake atau Juice berdasarkan rasio total Quantity
# ======================================================

mask_9 = base_mask & (df['Total Spent'] == 9.0)

df.loc[mask_9, 'Quantity'] = 3
df.loc[mask_9, 'Price Per Unit'] = 3.0

total_cake = df[df['Item'] == 'Cake']['Quantity'].sum()
total_juice = df[df['Item'] == 'Juice']['Quantity'].sum()

total_cake_juice = total_cake + total_juice

prob_cake = total_cake / total_cake_juice
prob_juice = total_juice / total_cake_juice

df.loc[mask_9, 'Item'] = np.random.choice(
    ['Cake', 'Juice'],
    size=mask_9.sum(),
    p=[prob_cake, prob_juice]
)


# ======================================================
# Total Spent = 16.0
# Quantity = 4, Price Per Unit = 4.0
# Item = Smoothie atau Sandwich berdasarkan rasio total Quantity
# ======================================================

mask_16 = base_mask & (df['Total Spent'] == 16.0)

df.loc[mask_16, 'Quantity'] = 4
df.loc[mask_16, 'Price Per Unit'] = 4.0

total_smoothie = df[df['Item'] == 'Smoothie']['Quantity'].sum()
total_sandwich = df[df['Item'] == 'Sandwich']['Quantity'].sum()

total_smoothie_sandwich = total_smoothie + total_sandwich

prob_smoothie = total_smoothie / total_smoothie_sandwich
prob_sandwich = total_sandwich / total_smoothie_sandwich

df.loc[mask_16, 'Item'] = np.random.choice(
    ['Smoothie', 'Sandwich'],
    size=mask_16.sum(),
    p=[prob_smoothie, prob_sandwich]
)


# ======================================================
# Cek jumlah data yang berhasil diisi
# ======================================================

print("Jumlah data Total Spent 25.0 yang diisi:", mask_25.sum())
print("Jumlah data Total Spent 9.0 yang diisi:", mask_9.sum())
print("Jumlah data Total Spent 16.0 yang diisi:", mask_16.sum())
```

Dengan cara yang sama, untuk data yang kosong di kolom Payment Method, akan digunakan random berdasarkan jumlah total dari setiap bagian pada kolom Payment Method.

```
# data yang kosong di kolom Payment Method akan diisi menggunakan random selection
# berdasarkan proporsi jumlah masing-masing Payment Method yang sudah ada

np.random.seed(42)

# Menentukan baris yang Payment Method-nya kosong
mask_payment = df['Payment Method'].isnull()

# Menghitung jumlah masing-masing metode pembayaran yang sudah ada
payment_counts = df['Payment Method'].value_counts()

print("Jumlah masing-masing Payment Method:")
print(payment_counts)

# Mengubah jumlah menjadi probabilitas
payment_probs = payment_counts / payment_counts.sum()

print("\nProbabilitas masing-masing Payment Method:")
print(payment_probs)

# Mengisi missing value berdasarkan probabilitas tersebut
df.loc[mask_payment, 'Payment Method'] = np.random.choice(
    payment_probs.index,
    size=mask_payment.sum(),
    p=payment_probs.values
)
```
